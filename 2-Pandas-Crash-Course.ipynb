Group BY 

```python
import pandas as pd
# Create dataframe
data1 = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB','Apple'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah','John'],
       'Sales':[200,120,340,124,243,350,400]}
```

```python
df = pd.DataFrame(data1)
df.head(7)
```

```python
by_comp = df.groupby("Company")
```

```python
print(by_comp.mean())
print("\n")
by_comp.std()
print("\n")
print(by_comp.min())
print("\n")
print(by_comp.max())
print("\n")
print(by_comp.count())
```

```python
by_comp.describe()
```
